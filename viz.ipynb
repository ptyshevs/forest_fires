```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/forestfires.csv')
```

```python
df.head()
# unfortunately X and Y are not real coordinates, so I cannot plot them directly
```

```python
sns.set(rc={'figure.figsize': (14, 8)})
```

```python
fig, ax = plt.subplots(ncols=2)
sns.boxplot(x='day', y='FFMC', hue='day', data=df, ax=ax[0])  # FFMC per day
sns.boxplot(x='month', y='FFMC', hue='month', data=df, ax=ax[1])
```

```python
df = df.drop(df[df.FFMC < 80].index)  # dropping outliers
```

```python
sns.heatmap(df.corr());
```

As we can see, no feature has correlation with the target variable (`area`)

```python
sns.lmplot('DC', 'area', df)  # linear regression shows that target variable skewed towards zero
```
